```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n_samples = 10000       # Number of sample means to generate
sample_size = 30        # Each sample's size
original_dist = np.random.uniform  # Underlying distribution

# Generate sample means
means = []
for _ in range(n_samples):
    sample = original_dist(low=0, high=1, size=sample_size)
    means.append(np.mean(sample))

means = np.array(means)

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(means, bins=100, density=True, alpha=0.6, label="Sample means (histogram)")

# Overlay normal PDF for comparison
mu_practice = np.mean(means)
sigma_practice = np.std(means)

mu_analyt = 1 / 2
sigma_analyt = np.sqrt(1/360)

x = np.linspace(mu_analyt - 4*sigma_analyt, mu_analyt + 4*sigma_analyt, 500)
plt.plot(x, norm.pdf(x, mu_analyt, sigma_analyt), 'r-', lw=2, label="Normal PDF")

plt.title("Central Limit Theorem Demonstration\n(Uniform[0, 1], Sample Size = 30)")
plt.text(0.65, 5, f'Real: $\mu$ = {mu_practice:.4f}, $\sigma$ = {sigma_practice:.4f}\nAnalytycal: $\mu$ = {mu_analyt:.4f}, $\sigma$ = {sigma_analyt:.4f}')

plt.xlabel("Sample mean")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
